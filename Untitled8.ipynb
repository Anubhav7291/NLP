```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputer.fit(df.iloc[:, 1:6])
df.iloc[:, 1:6] = imputer.transform(df.iloc[:, 1:6])
df.isna().any()
```

```python
imput = SimpleImputer(missing_values=np.nan, strategy='median')
imput.fit(df.iloc[:, 8:10])
df.iloc[:, 8:10] = imput.transform(df.iloc[:, 8:10])
df.isna().any()
```

```python
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp.fit(df.iloc[:,10:11])
df.iloc[:,10:11] = imp.transform(df.iloc[:,10:11])
df.isna().any()
```

```python
df['Dependents']=df['Dependents'].str.extract('(\d+)')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df.iloc[:,1] = le.fit_transform(df.iloc[:,1])
df.iloc[:,2] = le.fit_transform(df.iloc[:,2])
df.iloc[:,4] = le.fit_transform(df.iloc[:,4])
df.iloc[:,5] = le.fit_transform(df.iloc[:,5])
df.iloc[:,-1] = le.fit_transform(df.iloc[:,-1])
df.drop('Loan_ID',axis=1,inplace=True)
X=df
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df['ApplicantIncome']=df['ApplicantIncome']+df['CoapplicantIncome']
df['ApplicantIncome']=np.log(df['ApplicantIncome'])
df.drop('CoapplicantIncome',axis=1,inplace=True)
df
```

```python
df['LoanAmount']=df['LoanAmount']/df['Loan_Amount_Term']
df.drop('Loan_Amount_Term',axis=1,inplace=True)
df
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
X
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [-1])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
X
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(C= 0.25, intercept_scaling=0.25,random_state = 0)
classifier.fit(X,y)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(X)
cm = confusion_matrix(y, y_pred)
print(cm)
accuracy_score(y, y_pred)
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X, y = y, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {'C': [0.25,0.20,0.30,0.40,0.5, 0.75, 1],'penalty': ['l1', 'l2', 'elasticnet','none'],'intercept_scaling':[0.25,0.20,0.30,0.40,0.5, 0.75, 1]}
grid_search = GridSearchCV(estimator = classifier,
                           param_grid = parameters,
                           scoring = 'accuracy',
                           cv = 10,
                           n_jobs = -1)
grid_search.fit(X, y)
best_accuracy = grid_search.best_score_
best_parameters = grid_search.best_params_
print("Best Accuracy: {:.2f} %".format(best_accuracy*100))
print("Best Parameters:", best_parameters)
```
